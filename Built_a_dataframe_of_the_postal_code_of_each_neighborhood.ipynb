## Import library

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

## Define postal cost URL 

```python
url = "https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1012118802"
```

###Requests contents of the webpage in text format and store in a variable called data

```python
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data,"html5lib")
```

###Find all html tables in the web page

```python
tables = soup.find_all('table') # in html table is represented by the tag <table>
```

```python
len(tables)
```

### Find the name of table

```python
print(tables)
```

```python
for index,table in enumerate(tables):
    if ("wikitable sortable" in str(table)):
        table_index = index
print(table_index)
```

```python
print(tables[table_index].prettify())
```

##Tranfrom html. table to data frame

```python
neighborhoods_Toronto = pd.DataFrame(columns=["PastalCode", "Borough", "Neighbourhood",])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        pastalcode = col[0].text
        borough = col[1].text
        neighbourhood = col[2].text.strip()
        neighborhoods_Toronto = neighborhoods_Toronto.append({"PastalCode":pastalcode, "Borough":borough, "Neighbourhood":neighbourhood}, ignore_index=True)

neighborhoods_Toronto.head(5)
```

### Remove "\n" in Pastalcode and Borough column

```python
neighborhoods_Toronto.PastalCode = neighborhoods_Toronto.PastalCode.str.replace('\n', '')
neighborhoods_Toronto.Borough = neighborhoods_Toronto.Borough.str.replace('\n', '')
neighborhoods_Toronto.head(5)
```

##Drop Not assigned value 




```python
neighborhoods_Toronto = neighborhoods_Toronto.loc[neighborhoods_Toronto["Borough"] != "Not assigned"]
neighborhoods_Toronto.head()
```

## Reset index

```python
neighborhoods_Toronto.reset_index(drop= True,inplace=True)
neighborhoods_Toronto.head()
```
